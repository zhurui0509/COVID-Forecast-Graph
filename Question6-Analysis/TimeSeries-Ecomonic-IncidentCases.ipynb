```python
### read in the data ###
import pandas as pd
econ_aer_timeseries = pd.read_csv('timeseries_gps_retail_and_recreation.csv', index_col ='time_literal', parse_dates= True)
```

```python
econ_aer_timeseries.index.names = ['Date']
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,5))

ax1 = econ_aer_timeseries.groundtruth_value.plot(color='blue', grid=True, label='reported incident case (ground truth)')
ax2 = econ_aer_timeseries.val_econ.plot(color='red', grid=True, secondary_y=True, label='time spent at retail and recreation locations')

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()

plt.legend(h1+h2, l1+l2, loc=2)
#plt.xlabel('Time')

ax1.set_ylabel('number of reported incident case', color='b')
ax2.set_ylabel('time spent at retail and recreation locations \n relative to the January 2020 index period.', color='r')
plt.savefig('gps_retail_and_recreation_timeseries.png')
plt.show()
```
